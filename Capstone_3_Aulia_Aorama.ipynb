## **1.7 Metode Evaluasi Model**

Evaluasi model dilakukan menggunakan beberapa metrik regresi untuk memperoleh gambaran performa secara menyeluruh dan tidak bergantung pada satu indikator saja.

> **Mean Squared Error (MSE)** digunakan sebagai metrik utama untuk **melihat tingkat kesalahan prediksi model dan memastikan bahwa nilai error yang dihasilkan serendah mungkin**. Metrik ini mengukur rata-rata kuadrat selisih antara harga aktual dan harga prediksi, sehingga memberikan penalti yang lebih besar terhadap kesalahan yang signifikan. Dengan menjadikan MSE sebagai metrik utama, proses pemilihan model difokuskan pada upaya meminimalkan kesalahan prediksi secara keseluruhan pada seluruh data kendaraan.

> Selanjutnya, **Root Mean Squared Error (RMSE) digunakan sebagai metrik pendukung untuk menginterpretasikan besarnya kesalahan dalam satuan harga yang sama dengan variabel target**. Jika MSE memberikan nilai dalam bentuk kuadrat, maka RMSE mengembalikannya ke satuan moneter, sehingga lebih mudah dipahami dalam konteks bisnis, misalnya untuk mengetahui rata-rata selisih harga prediksi dalam mata uang aktual.

> Selain itu, **R-squared (R²) digunakan untuk mengukur seberapa besar variasi harga mobil bekas yang dapat dijelaskan oleh fitur-fitur dalam model**. Nilai R² memberikan gambaran mengenai kemampuan model dalam menangkap pola hubungan antara karakteristik kendaraan dan harga. Semakin tinggi nilai R², semakin baik model dalam menjelaskan variasi harga di pasar.

> Sebagai pelengkap, **Mean Absolute Percentage Error (MAPE) digunakan untuk melihat kesalahan prediksi dalam bentuk persentase. Metrik ini membantu interpretasi dari sudut pandang bisnis, karena dapat menunjukkan rata-rata persentase kesalahan harga yang dihasilkan model**. Dengan kombinasi keempat metrik tersebut, evaluasi model dilakukan secara menyeluruh agar model yang dipilih tidak hanya memiliki error rendah secara statistik, tetapi juga relevan dan dapat diinterpretasikan dalam konteks pasar mobil bekas.

### **3.2.2. Outlier Data Numerik**

```python
plt.figure(figsize=(13, 10))
plotnumber = 1
numerik = df_eda.select_dtypes(include='number').columns

for feature in numerik:
    ax = plt.subplot(5, 3, plotnumber)
    sns.boxplot(data=df_eda,x=feature)
    plt.title(feature, fontsize=20)
    plotnumber += 1
    plt.tight_layout()
```

```python
# Menggunakan IQR untuk mendapatkan jumlah outliers

hasil = []

for kolom in numerik:
    q1 = df_eda[kolom].quantile(0.25)
    q3 = df_eda[kolom].quantile(0.75)
    iqr = q3 - q1
    whis = 1.5 * iqr

    upper_bound = q3 + whis
    lower_bound = q1 - whis

    jumlah_outliers = len(df_eda[(df_eda[kolom] < lower_bound) | (df_eda[kolom] > upper_bound)])

    hasil.append({
        'Kolom': kolom,
        'Q1': q1,
        'Q3': q3,
        'IQR': iqr,
        'Lower_Bound': lower_bound,
        'Upper_Bound': upper_bound,
        'Jumlah_Outliers': jumlah_outliers
    })
df_outliers = pd.DataFrame(hasil)
df_outliers
```

### **3.2.3. Cek Normalitas Data Numerik**

Tes normalitas digunakan untuk penentuan penggunaan uji korelasi dan penentuan scaling pada data preprosesing.

D’Agostino & Pearson test digunakan karena ukuran sampel dalam penelitian ini besar (>5.000 observasi). Pada sampel besar, uji Shapiro–Wilk cenderung terlalu sensitif dan hampir selalu menghasilkan hasil signifikan meskipun penyimpangannya kecil.

D’Agostino & Pearson lebih sesuai karena menguji normalitas berdasarkan skewness dan kurtosis, sehingga memberikan evaluasi yang lebih stabil dan representatif terhadap bentuk distribusi data.


```python
from scipy.stats import normaltest

listPvalue = []
listStatusNormal = []
listKolom = ['Car_Age','Engine_Size','Mileage','Price']

for i in listKolom:
    stat, pvalue = normaltest(df_eda[i])

    if pvalue <= 0.05:
        status = 'Distribusi Data Tidak Normal'
    else:
        status = 'Distribusi Data Normal'

    listPvalue.append(pvalue)
    listStatusNormal.append(status)

Normalitas = pd.DataFrame({
    'Column Name': listKolom,
    'Pvalue': listPvalue,
    'Status Normalitas': listStatusNormal
})

Normalitas
```

**Justifikasi Pemilihan Variabel Car_Age dibandingkan Year**

Dalam proses seleksi fitur, variabel **Year** dan **Car_Age** dievaluasi karena keduanya merepresentasikan informasi temporal terkait kendaraan. Secara matematis, kedua variabel memiliki hubungan deterministik:

[
Car_Age = Current_Year - Year
]

Hubungan linear sempurna tersebut menyebabkan keduanya mengandung informasi yang identik. Oleh karena itu, penggunaan keduanya secara simultan dalam model regresi berpotensi menimbulkan **multicollinearity**, yaitu kondisi ketika dua atau lebih variabel independen memiliki korelasi tinggi. Multicollinearity dapat menyebabkan ketidakstabilan estimasi koefisien, meningkatkan varians parameter, serta menurunkan interpretabilitas model.

Berdasarkan pertimbangan tersebut, hanya satu variabel yang dipertahankan dalam proses modeling.

1. Pertimbangan Temporal Distribution Shift

Model yang dikembangkan dirancang untuk digunakan secara berulang tanpa retraining tahunan. Jika menggunakan variabel **Year**, distribusi data akan terus berubah seiring munculnya kendaraan dengan tahun produksi yang lebih baru. Kondisi ini berpotensi menyebabkan **temporal distribution shift**, yaitu perubahan distribusi fitur antara data pelatihan dan data penggunaan aktual.

Sebagai contoh, jika model dilatih dengan data hingga tahun 2023, maka pada saat digunakan di tahun 2027 model akan menerima nilai Year yang belum pernah diamati sebelumnya. Hal ini dapat menurunkan performa prediksi karena model dipaksa melakukan ekstrapolasi di luar distribusi pelatihan.

Sebaliknya, variabel **Car_Age** merepresentasikan usia kendaraan relatif terhadap waktu prediksi dilakukan. Dengan menghitung Car_Age secara dinamis, model tetap memproses informasi dalam skala yang konsisten, sehingga lebih robust terhadap perubahan kalender waktu.

---

2. Representasi Ekonomi yang Lebih Relevan

Harga mobil bekas secara ekonomi mengikuti pola depresiasi berdasarkan umur kendaraan, bukan berdasarkan angka tahun absolut. Depresiasi umumnya terjadi secara signifikan pada tahun-tahun awal penggunaan dan cenderung melandai setelah periode tertentu. Oleh karena itu, variabel Car_Age secara langsung mencerminkan mekanisme pembentukan harga dalam pasar mobil bekas.

---

3. Efisiensi dan Reduksi Redundansi Fitur

Karena Year dan Car_Age membawa informasi yang sama, mempertahankan keduanya tidak memberikan peningkatan informasi (no additional predictive signal), melainkan hanya menambah kompleksitas model. Dengan memilih satu representasi yang lebih stabil dan ekonomis, model menjadi lebih efisien dan lebih mudah diinterpretasikan.

---

**Kesimpulan**

Berdasarkan pertimbangan:

1. Potensi multicollinearity,
2. Risiko temporal distribution shift,
3. Kesesuaian dengan teori depresiasi kendaraan,
4. Kebutuhan stabilitas model dalam penggunaan jangka panjang,

variabel **Car_Age** dipilih sebagai fitur dalam proses pemodelan, sementara variabel **Year** tidak digunakan secara langsung dalam training model namun tetap dipertahankan untuk keperluan pelaporan dan penyajian hasil kepada pengguna.

### **3.2.7. Hubungan Kategori Vs Price**

Mengingat distribusi harga tidak normal dan mengandung outlier signifikan, digunakan uji non-parametrik Kruskal-Wallis untuk mengevaluasi perbedaan harga antar kategori. Pendekatan ini dipilih karena lebih robust terhadap pelanggaran asumsi normalitas dan heterogenitas varians.

```python
from scipy.stats import kruskal
import pandas as pd

categorical_cols = df_eda.select_dtypes(include='object').columns
results = []

for col in categorical_cols:

    groups = [group["Price"].values
              for name, group in df.groupby(col)]

    if len(groups) > 1:
        stat, p_value = kruskal(*groups)
        results.append([col, stat, p_value])

kruskal_df = pd.DataFrame(results,
                          columns=["Feature", "H_stat", "p_value"])

# Tambahkan interpretasi
def interpretasi(row):
    if row["p_value"] < 0.05:
        return "Terdapat perbedaan harga yang signifikan antar kategori"
    else:
        return "Tidak terdapat perbedaan harga yang signifikan"

kruskal_df["Interpretasi"] = kruskal_df.apply(interpretasi, axis=1)

kruskal_df = kruskal_df.sort_values("H_stat", ascending=False)

kruskal_df
```

Hasil uji menunjukkan bahwa **Type memiliki pengaruh paling kuat terhadap variasi harga mobil**, karena memiliki nilai H-stat tertinggi (976.76). Ini berarti distribusi harga antar tipe mobil berbeda sangat signifikan.

# **5. Data Preprocessing**

```python
# Membuat transformer untuk mengubah Year menjadi Car_Age
# dan menghapus kolom Year agar tidak terjadi multicollinearity

from sklearn.base import BaseEstimator, TransformerMixin
from datetime import datetime
class CarAgeTransformer(BaseEstimator, TransformerMixin):

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        current_year = datetime.now().year
        X["Car_Age"] = current_year - X["Year"]
        X = X.drop(columns=["Year"])
        return X
```

```python
!pip install category_encoders

# Library untuk preprocessing dan encoding
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from category_encoders import BinaryEncoder
from sklearn.ensemble import RandomForestRegressor
# Fitur numerik (akan di-scale)
numeric_features = ["Mileage", "Car_Age", "Engine_Size"]

# Kategori dengan cardinality kecil (<=5 kategori)
low_card_cat = ["Gear_Type", "Origin", "Options", "Negotiable"]

# Kategori dengan cardinality besar
high_card_cat = ["Type", "Make", "Region"]

# Membuat preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ("num", RobustScaler(), numeric_features),
        ("low_cat", OneHotEncoder(handle_unknown="ignore"), low_card_cat),
        ("high_cat", BinaryEncoder(), high_card_cat),
    ],
    remainder="drop"
)
model = Pipeline(steps=[
    ("car_age", CarAgeTransformer()),   # Transform Year → Car_Age
    ("preprocessing", preprocessor),    # Encoding & Scaling
    ("regressor", RandomForestRegressor(random_state=42))
])
```

# **8. Cross Validation**

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import VotingRegressor, StackingRegressor, BaggingRegressor, RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor

# ============================================================
# DEFINE MODELS
# ============================================================

lr = LinearRegression()
lasso = Lasso(alpha=0.0005, random_state=2024)
ridge = Ridge(random_state=2024)
dt = DecisionTreeRegressor(random_state=2024)
knn = KNeighborsRegressor()
rf = RandomForestRegressor(random_state=2024, n_jobs=-1)

# Ensemble tambahan
voting = VotingRegressor(
    estimators=[
        ('lasso', lasso),
        ('tree', dt),
        ('knn', knn)
    ]
)

stacking = StackingRegressor(
    estimators=[
        ('lasso', lasso),
        ('tree', dt),
        ('knn', knn)
    ],
    final_estimator=Ridge()
)

bagging = BaggingRegressor(
    estimator=knn,
    random_state=2024
)

ada = AdaBoostRegressor(random_state=2024)
gboost = GradientBoostingRegressor(random_state=2024)
xgb = XGBRegressor(random_state=2024, n_jobs=-1)

# Dictionary model
models = {
    "LinearRegression": lr,
    "Lasso": lasso,
    "Ridge": ridge,
    "DecisionTree": dt,
    "KNN": knn,
    "RandomForest": rf,
    "Voting": voting,
    "Stacking": stacking,
    "Bagging": bagging,
    "AdaBoost": ada,
    "GradientBoosting": gboost,
    "XGBoost": xgb
}
```

```python
# ============================================================
# DEFINE SCORING
# ============================================================

scoring = {
    "MSE": "neg_mean_squared_error",
    "RMSE": "neg_root_mean_squared_error",
    "MAE": "neg_mean_absolute_error",
    "MAPE": "neg_mean_absolute_percentage_error"
}
```

```python
# ============================================================
# CROSS VALIDATION
# ============================================================

from sklearn.model_selection import cross_validate, KFold
from sklearn.pipeline import Pipeline

results = []

for name, algo in models.items():

    pipe_model = Pipeline([
        ("car_age", CarAgeTransformer()),
        ("preprocessing", preprocessor),
        ("modeling", algo)
    ])

    cross_val = cross_validate(
        estimator=pipe_model,
        X=X_train,
        y=y_train,
        cv=5,
        scoring=scoring,
        n_jobs=-1
    )

    # Ambil skor dan ubah ke positif
    mse_scores = -cross_val["test_MSE"]
    rmse_scores = -cross_val["test_RMSE"]
    mae_scores = -cross_val["test_MAE"]
    mape_scores = -cross_val["test_MAPE"]

    results.append({
        "Model": name,

        # MSE
        "MSE_score": mse_scores.round(2).tolist(),
        "MSE_mean": round(mse_scores.mean(), 2),
        "MSE_std": round(mse_scores.std(), 2),

        # RMSE
        "RMSE_score": rmse_scores.round(2).tolist(),
        "RMSE_mean": round(rmse_scores.mean(), 2),
        "RMSE_std": round(rmse_scores.std(), 2),

        # MAE
        "MAE_score": mae_scores.round(2).tolist(),
        "MAE_mean": round(mae_scores.mean(), 2),
        "MAE_std": round(mae_scores.std(), 2),

        # MAPE
        "MAPE_score": mape_scores.round(4).tolist(),
        "MAPE_mean": round(mape_scores.mean(), 4),
        "MAPE_std": round(mape_scores.std(), 4)
    })

df_cv_full = (
    pd.DataFrame(results)
    .sort_values("MSE_mean")  # karena MSE metrik utama
    .reset_index(drop=True)
)

df_cv_full
```

# **9. Hyperparameter Tunning**

```python
from sklearn.model_selection import RandomizedSearchCV

param_xgb = {
    "modeling__n_estimators": [100, 200, 300],
    "modeling__max_depth": [3, 5, 7, 9, 12],
    "modeling__learning_rate": [0.01, 0.05, 0.1],
    "modeling__subsample": [0.8, 1.0],
    "modeling__colsample_bytree": [0.8, 1.0],
    "modeling__reg_alpha": [0, 0.01, 0.1],
    "modeling__reg_lambda": [1, 5, 10]
}

pipe_xgb = Pipeline([
    ("car_age", CarAgeTransformer()),   # Transform Year → Car_Age
    ("preprocessing", preprocessor),    # Encoding & Scaling
    ("modeling", XGBRegressor(random_state=42, n_jobs=-1))
])

# Use the 'scoring' dictionary defined earlier for multiple metrics
random_search_xgb = RandomizedSearchCV(
    estimator=pipe_xgb,
    param_distributions=param_xgb,
    n_iter=50,               # jumlah kombinasi dicoba
    scoring=scoring,         # Updated to use the 'scoring' dictionary
    cv=5,
    verbose=1,
    n_jobs=-1,
    random_state=42,
    refit='MSE' # Specify which metric to use for refitting the best estimator (use the key from the scoring dict)
)

random_search_xgb.fit(X_train, y_train)
```


XGBoost (Extreme Gradient Boosting) adalah algoritma berbasis **decision tree ensemble** yang bekerja dengan teknik **boosting**.

Cara kerjanya secara sederhana:

1. Model membangun **pohon keputusan pertama** untuk memprediksi harga.
2. Model menghitung **error (selisih prediksi dengan harga asli)**.
3. Pohon berikutnya dibangun untuk **memperbaiki error sebelumnya**.
4. Proses ini diulang berkali-kali.
5. Hasil akhir adalah gabungan dari semua pohon yang secara bertahap memperbaiki kesalahan.

Intinya:

> Setiap pohon baru belajar dari kesalahan pohon sebelumnya.

XGBoost menggunakan:

* Gradient Descent untuk meminimalkan loss function
* Regularization untuk mencegah overfitting
* Optimisasi komputasi agar lebih cepat dan efisien

---

**Kenapa XGBoost Cocok untuk Kasus Ini?**

karena **prediksi harga mobil bekas**, yang memiliki karakteristik:

1. Hubungan Non-Linear

Harga tidak naik atau turun secara linear terhadap usia, mesin, atau mileage.
XGBoost sangat baik menangkap pola non-linear.

2. Banyak Fitur Kategorikal

Seperti Type, Make, Region.
Tree-based model seperti XGBoost menangani fitur encoded dengan sangat baik.

3. Interaksi Antar Variabel

Contoh:

* Mobil tua + mesin besar → bisa tetap mahal
* Make tertentu + fitur lengkap → harga melonjak

XGBoost mampu menangkap interaksi kompleks antar fitur.

4. Data dengan Skala Berbeda

Tree model tidak sensitif terhadap scaling seperti model linear.

5. Robust terhadap Outlier Moderat

Lebih stabil dibanding regresi linear biasa.

---

**Kenapa Performanya Lebih Baik?**

Karena:

* Model linear hanya menangkap hubungan lurus.
* Random Forest kuat, tetapi tidak sequentially memperbaiki error.
* XGBoost memperbaiki error secara bertahap dan terkontrol.

Itu sebabnya:

* MSE turun signifikan
* R² meningkat
* Model lebih stabil
